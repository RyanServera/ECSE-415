## 2.2 Laplacian of Gaussian

```python
# Useful imports
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image of a woman as a grayscale image
woman_path = "./woman.png"
woman = cv2.imread(woman_path, cv2.IMREAD_GRAYSCALE)

# Created a new image which has a laplacian edge detector applied to it
woman_laplacian = cv2.Laplacian(woman, cv2.CV_32F, ksize=5)

# Takes the height and width of the picture
height, width = woman_laplacian.shape

# Inializes a blank image with the same width and height as the 
# original image
woman_thresh = np.zeros((height, width), dtype=np.uint8)

# If the intensity of the pixel is higher then 700 insert a white
# point in the woman_thresh image
for row in range(0, height):
    for col in range(0, width):
        if woman_laplacian[row,col] > 700:
            woman_thresh[row, col] = 255

# Plots the original, laplacian and thresholded image
plt.figure(figsize=(20, 20))
plt.subplot(131), plt.imshow(woman , cmap="gray")
plt.title("Original"), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(woman_laplacian, cmap="gray")
plt.title("Laplacian"), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(woman_thresh, cmap="gray")
plt.title(":Laplacian Threshold"), plt.xticks([]), plt.yticks([])
plt.show()
```
