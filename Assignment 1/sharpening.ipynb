# 1.2 Sharpening

## 1. Sharpening Rice

```python
# Useful imports
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image of rice as a grayscale image
rice_path = "./rice.png"
rice = cv2.imread(rice_path, cv2.IMREAD_GRAYSCALE)

# Creates a 5x5 box kernel
kernel = np.ones((5,5), np.float32)/(5*5)

# Creates a blured image using a box filter
box_rice = cv2.filter2D(rice, -1, kernel)

# Sharpens the image by adding details
details1 = rice - box_rice
sharpened_rice1 = rice + details1

# Plots the original image and the sharpened image
plt.figure(figsize=(20, 20))
plt.subplot(131), plt.imshow(rice,  cmap="gray")
plt.title("Original"), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(details1,  cmap="gray")
plt.title("Details"), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(sharpened_rice1)
plt.imshow(sharpened_rice1,  cmap="gray")
plt.title("Sharpened"), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Creates a blured image using a Gaussian Filter (5x5, variance 2)
gaussian_rice = cv2.GaussianBlur(rice, (5,5), 2**(1/2), 2**(1/2))

# Sharpens the image by adding the details
details2 = rice - gaussian_rice
sharpened_rice2 = rice + details2

# Plots the original image and the sharpened rice
plt.figure(figsize=(20, 20))
plt.subplot(131), plt.imshow(rice, cmap="gray")
plt.title("Original"), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(details2, cmap="gray")
plt.title("Details"), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(sharpened_rice2, cmap="gray")
plt.title("Sharpened"), plt.xticks([]), plt.yticks([])
plt.show()
```

## 2. Commenting on the sharpening

```python
plt.figure(figsize=(20, 20))
plt.subplot(131), plt.imshow(rice, cmap="gray")
plt.title("Original"), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(sharpened_rice1, cmap="gray")
plt.title("Sharpened Box"), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(sharpened_rice2, cmap="gray")
plt.title("Sharpened Gaussian"), plt.xticks([]), plt.yticks([])
plt.show()
```

#### The Gaussian Filter is supposed give the better result because it is more selective, therefore the details added will be more defined in terms of contrast than the box one.

#### However, the results in practice are not noticeably different.
